### Starting clean up
The following cleaning steps were chosen to handle missing data, remove duplicates, fix data types, and reduce the impact of outliersâ€”ensuring the dataset was accurate, consistent, and ready for analysis.

```python
# Check missing values per column to understand data gaps
print("Step 1: Missing values per column:")
print(df.isnull().sum())
```

```python
import numpy as np

# Identify numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Fill missing numeric values with median of each column
print("\nStep 2: Filling missing numeric values with median")
for col in numeric_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    print(f"Filled '{col}' missing values with median: {median_val}")
```

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Fill missing categorical values with mode of each column
print("\nStep 3: Filling missing categorical values with mode")
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
    print(f"Filled '{col}' missing values with mode: {mode_val}")
```

```python
# Calculate percentage of missing values per column
missing_pct = df.isnull().mean()

# Identify columns with more than 50% missing data
cols_to_drop = missing_pct[missing_pct > 0.5].index

# Drop these columns if any found
if len(cols_to_drop) > 0:
    print(f"\nStep 4: Dropping columns with >50% missing data: {list(cols_to_drop)}")
    df = df.drop(columns=cols_to_drop)
else:
    print("\nStep 4: No columns to drop based on missingness > 50%")
```

```python
# Count duplicates
num_duplicates = df.duplicated().sum()
print(f"\nStep 5: Number of duplicate rows: {num_duplicates}")

# Remove duplicates
df = df.drop_duplicates()
print("Duplicates removed.")
```

```python
# Check duplicates based on country_name and year
dup_key_count = df.duplicated(subset=['country_name', 'year']).sum()
print(f"\nStep 6: Duplicates based on 'country_name' and 'year': {dup_key_count}")
```

```python
from scipy import stats
import numpy as np

print("\nStep 8: Detecting and capping outliers in numeric columns")

for col in numeric_cols:
    # Calculate z-scores
    z_scores = np.abs(stats.zscore(df[col]))
    outliers_count = (z_scores > 3).sum()
    print(f"Outliers detected in '{col}': {outliers_count}")

    # Cap values at 1st and 99th percentiles
    lower_bound = df[col].quantile(0.01)
    upper_bound = df[col].quantile(0.99)

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
```

```python
print("\nStep 9: Ensuring correct data types")

# Convert 'year' to integer
df['year'] = df['year'].astype(int)
print("'year' converted to int")

# Convert categorical columns to category dtype for efficiency
for col in categorical_cols:
    df[col] = df[col].astype('category')
print("Categorical columns converted to 'category' dtype")
```

```python
# Plot distributions
plt.figure(figsize=(15,10))
df[['Inflation (CPI %)', 'GDP (Current USD)', 'GDP per Capita (Current USD)']].hist()
plt.tight_layout()
plt.show()
```
